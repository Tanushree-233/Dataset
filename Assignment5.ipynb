```python
import pandas as pd

# Load CSV
df = pd.read_csv("train.csv")

# Now you can safely run:
drop_cols = [col for col in df.columns if df[col].isnull().mean() > 0.5]
df.drop(columns=drop_cols, inplace=True)

print("Dropped columns:", drop_cols)
```

```python
# Fill missing values
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("None")  # for categorical columns
    else:
        df[col] = df[col].fillna(df[col].median())  # for numerical columns

print(" Missing values filled.")
```

```python
# Apply one-hot encoding to categorical columns
df = pd.get_dummies(df)

print(" One-hot encoding complete. Final shape:", df.shape)
```

```python
df.head()
```

```python
df['TotalBathrooms'] = df['FullBath'] + 0.5 * df['HalfBath']
```

```python
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
```

```python
df['HouseAge'] = df['YrSold'] - df['YearBuilt']
```

```python
df['GarageAge'] = df['YrSold'] - df['GarageYrBlt']
df['GarageAge'].fillna(df['GarageAge'].median(), inplace=True)
```
